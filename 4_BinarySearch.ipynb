7-1. 순차 탐색

```python
# 순차 탐색 소스코드 구현 (데이터 개수가 N개일 때 최대 N번의 비교 연산이 필요하므로 시간 복잡도는 O(N))

def sequential_search(n, target, array):
    # 각 원소를 하나씩 확인
    for i in range(n):
        # 현재의 원소가 찾고자 하는 원소와 동일한 경우
        if array[i] == target:
            print('찾고자 하는 문자열의 위치')
            return i + 1 # 현재의 위치 반환
        
print('생성할 원소 개수를 입력한 다음, 한 칸 띄우고 찾을 문자열을 입력하세요!')
input_data = input().split()

n = int(input_data[0]) # 원소 개수
target = input_data[1] # 찾고자 하는 문자열

print('찾고자 하는 문자열 :', target)

print('앞서 적은 원소 개수만큼 문자열을 입력하세요!')
array = input().split()

print('입력 문자열 :', array)

# 순차 탐색 수행 결과
print(sequential_search(n, target, array))
```

7-2. 재귀 함수로 구현한 이진 탐색

```python
def binary_serach(array, target, start, end):
    if start > end:
        return None
    
    mid = (start + end) // 2
    # 찾은 경우 중간점 인덱스 반환
    if array[mid] == target:
        return mid
    
    # 중간점의 값보다 찾고자 하는 값이 작은 경우 왼쪽 확인
    elif array[mid] > target:
        return binary_serach(array, target, start, mid - 1)
    
    # 중간점의 값보다 찾고자 하는 값이 큰 경우 오른쪽 확인
    else:
        return binary_serach(array, target, mid+1, end)
    
# n(원소의 개수)과 targer(찾고자 하는 문자열) 입력
n, target = list(map(int, input('원소의 개수, 찾고자 하는 문자열 입력: ').split()))

# 전체 원소 입력
array = list(map(int, input('전체 원소 입력: ').split()))

# 이진 탐색 수행 결과 출력
result = binary_serach(array, target, 0, n-1)
if result == None:
    print('원소가 존재하지 않습니다.')
else:
    print(result + 1)
```

7-3. 반복문으로 구현한 이진 탐색 

```python
def binary_serach(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2
        # 찾은 경우 중간점 인덱스 반환
        
        if array[mid] == target:
            return mid
        
        # 중간점의 값보다 찾고자 하는 값이 적은 경우 왼쪽 확인
        elif array[mid] > target:
            end = mid - 1
            
        # 중간점의 값보다 찾고자 하는 값이 큰 경우 오른쪽 확인
        else:
            start = mid + 1
            
    return None

# n(원소의 개수)과 targer(찾고자 하는 문자열) 입력
n, target = list(map(int, input('원소의 개수, 찾고자 하는 문자열 입력: ').split()))

# 전체 원소 입력
array = list(map(int, input('전체 원소 입력: ').split()))

# 이진 탐색 수행 결과 출력
result = binary_serach(array, target, 0, n-1)
if result == None:
    print('원소가 존재하지 않습니다.')
else:
    print(result + 1)
```

## 실전 문제2. 부품 찾기

전자 매장에 부품이 N개 있다.<br>
각 부품은 정수 형태의 고유한 번호가 있다.<br>
어느날 손님이 M개의 종류의 부품을 대량으로 구매하겠다며 당일 날 견적서를 요청했다.<br>
주인은 때를 놓치지 않고 손님이 문의한 부품 M개의 종류를 모두 확인해서 견적서를 작성해야 한다.<br>
이때 가게 안에 부품이 모두 있는지 확인하는 프로그램을 작성해보자.

#### 이진 탐색 풀이 

```python
def binary_search(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2
        
        if array[mid] == target:
            return mid
        elif array[mid] > target:
            end = mid - 1
        else:
            start = mid + 1
    return None

n = int(input('가게 안 부품 수 입력: ')) # 가게 안 부품 수
array = list(map(int, input('가게 안 부품 번호 입력: ').split())) # 가게 안 부품 번호
array.sort() # 이진 탐색을 수행하기 위해 정렬

print('가게 안 부품 번호', array)

m = int(input('손님이 요구한 부품 수 입력: ')) # 손님이 요구한 부품 수
request = list(map(int, input('손님이 요청한 부품 번호 입력: ').split())) # 손님이 요청한 부품 번호

print('손님이 요청한 부품 번호', request)

# 손님이 확인 요청한 부품 번호를 하나씩 확인
for i in request:
    result = binary_search(array, i, 0, n-1)
    # 해당 부품이 존재하는지 확인
    if result != None:
        print('yes', end=' ')
    else:
        print('no', end=' ')
```

## 실전 문제3. 떡볶이 떡 만들기

오늘은 떡볶이 떡을 만드는 날! 한 봉지 안에 들어가는 떡의 총 길이는 절단기로 잘라서 맞춰준다.<br>
절단기에 높이(H)를 지정하면 줄지어진 떡을 한 번에 절단한다.<br>
높이가 H보다 긴 떡은 H 위의 부분이 잘릴 것이고, 낮은 떡은 잘리지 않는다.<br>
손님이 왔을 때 요청한 총 길이가 M일 때 적어도 M만큼의 떡을 얻기 위해 절단기에 설정할 수 있는 높이의 최댓값을 구하는 프로그램을 작성하시오.

```python
# 적절한 높이를 찾을 때까지 절단기의 높이 H를 반복해서 조정
# '현재 이 높이로 자르면 조건을 만족할 수 있는가?'를 확인한 뒤에 조건의 만족 여부(예 혹인 아니오)에 따라서 탐색 범위를 좁혀서 해결
n, m = map(int, input('떡의 개수, 요청한 떡의 길이 각각 입력: ').split(' '))
height = list(map(int, input('떡의 개별 높이: ').split()))

# 이진 탐색을 위한 시작점과 끝점 설정
start = 0
end = max(height)

# 이진 탐색 수행(반복적)
result = 0
while (start <= end):
    total = 0
    mid = (start + end) // 2
    for i in height:
        # 잘랐을 때 떡의 양 계산
        if i > mid:
            total += i - mid
            
    # 떡의 양이 부족한 경우 더 많이 자르기
    if total < m:
        end = mid - 1
    
    # 떡의 양이 충분한 경우 덜 자르기
    else:
        result = mid # 최대한 덜 잘랐을 때가 정답
        start = mid + 1
        
print(result)
```
